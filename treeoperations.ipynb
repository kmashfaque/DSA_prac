```python
# Level Finding 

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def findLevel(root, key, level = 1):
    
    if not root:
        return 0
    if root.key == key:
        return level

    left_subtree = findLevel(root.left, key, level + 1)
    if left_subtree != 0:
        return left_subtree

    return findLevel(root.right, key, level + 1)


root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.left.right = Node(50)
root.left.left.left = Node(60)

print("Level of node 50 is:", findLevel(root, 50))
```

```python
# Tree Height

def findDepth(root):
    if root is None:
        return 0

    return 1 + max(findDepth(root.left), findDepth(root.right))

findDepth(root)
```

```python
# Find Depth

def findDepth(root, key, depth = 0):
    if not root:
        return -1
    if root.key == key:
        return depth
    
    left = findDepth(root.left, key, depth + 1)
    if left != -1:
        return left

    return findDepth(root.right, key, depth + 1)

findDepth(root, 60)
```
