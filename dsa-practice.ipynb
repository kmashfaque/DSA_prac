# TWO SUM Problem

```python

def twoSum(num, target):
    hashMap = {} # val : index

    for i, n in enumerate(num):
        diff = target - n 

        if diff in hashMap:
            return [diff, num[i]]
        hashMap[n] = i
    return

    
num = [1, 2, 3, 4, 5, 6, 7, 9, 12]
twoSum(num, 12)
```

# Two Sum with Sorted Array (Check again)

```python
def twoSumSorted(arr, target):
    res = []
    l, r = 0, len(arr) - 1

    while l < r:
        current_sum = arr[l] + arr[r]
        
        if current_sum > target:
            r -= 1
        elif current_sum < target:
            l+=1
        else:
            res.extend([l, r])
            return res
    

arr = [1,2,3,4,5]
twoSumSorted(arr, 10)
            
    
```

# Merge Sorted Array

```python
def mergeSorted_array(num1, m, num2, n):
    # last in num1
    last = m + n -1

    while m > 0 and n > 0:
        
        if num1[m - 1] > num2[n - 1]:
            num1[last] = num1[m - 1]
            m -= 1
        else:
            num1[last] = num2[n - 1]
            n -= 1
        last -= 1

    while n > 0:
        num1[last] = num2[n - 1]
        n, last = n - 1, last - 1

    while m > 0:
        num1[last] = num1[m - 1]
        m, last = m - 1, last - 1

    return num1
            

num1 = [1, 2, 3, 0, 0, 0]
m = 3
num2 = [2,5,6]
n = 3
mergeSorted_array(num1, m, num2, n)
```

```python
def maxSubArr(vals):
    currSum = 0
    Max_sub = vals[0]

    for n in vals:
        if currSum < 0:
            currSum = 0
        currSum += n
        Max_sub = max(currSum, Max_sub)
        
    return Max_sub

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
maxSubArr(arr)
    
```

# Product of array except Self(Sufffix and Prefix) (Check again)

```python
def product(nums):
    res = [1] * len(nums)
    prefix = 1
    for i in range(len(nums)):
        res[i] = prefix
        prefix *= nums[i]
               
    postfix = 1
    for i in range(len(nums)-1, -1, -1):
         res[i] *= postfix 
         postfix *=nums[i]
        
    return res
        

nums = [1,2,3,4, 5]
product(nums)
```

# Find Minimum in Rotated Sorted Array(Binary Search)

```python
def findMin(num):
    res = num[0]
    l,r = 0, len(num)-1 
    while l<=r:
        # checks whether array elements are in the acending order
        # if it is in ascending order then we will break the loop.
        # as we have found the minimum element
        if num[l] < num[r]:
            res = min(res, num[l])
            break
        m = (l+r)//2
        res = min(res, num[m])
        if num[m] > num[l]:
            l = m + 1
        else:
            r = m - 1
    return res

rotate = [3,4,5,6,7,8,9, 1,2]
findMin(rotate)
```

# Sorting an Array

```python
def sortArray(nums):

    def merge(arr, L, M, R):
        # Create temp subarrays
        left = arr[L:M+1]
        right = arr[M+1:R+1]

        i, j, k = L, 0, 0

        # Merge the two sorted halves
        while j < len(left) and k < len(right):
            if left[j] <= right[k]:
                arr[i] = left[j]
                j += 1
            else:
                arr[i] = right[k]
                k += 1
            i += 1

        # Copy remaining elements of left[]
        while j < len(left):
            arr[i] = left[j]
            j += 1
            i += 1

        # Copy remaining elements of right[]
        while k < len(right):
            arr[i] = right[k]
            k += 1
            i += 1

    def mergeSort(arr, l, r):
        if l == r:
            return arr
        m = (l + r) // 2
        mergeSort(arr, l, m)
        mergeSort(arr, m + 1, r)
        merge(arr, l, m, r)

    mergeSort(nums, 0, len(nums) - 1)
    return nums


# ✅ Example usage
arr = [1, 2, 3, 4, 6, 8, 5, 8, 9, 10, 17, 16, 14, 7,11,12]
print(sortArray(arr))
```

# Rotating An Array

```python
def rotate(arr, k):
    k = k % len(arr)
    l, r = 0, len(arr) - 1
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l, r = l+1, r-1
    l, r = 0, k-1
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l, r = l+1, r-1 
    l, r = k, len(arr) - 1
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l, r = l+1, r-1
    return arr


        
arr = [1, 2, 3, 4, 5, 6, 7, 8, 8, 9, 10, 11, 12, 14, 16, 17]
rotate(arr, 3)
```

# Container with Most Water

```python
def containerMaxWater(arr):
    l, r = 0, len(arr) - 1
    max_area = 0

    while l < r:
        height = min(arr[l], arr[r])
        width = r - l
        max_area = max(max_area, height * width)

        # Move the pointer pointing to the shorter line
        if arr[l] < arr[r]:
            l += 1
        else:
            r -= 1

    return max_area

arr = [1,8,6,2,5,4,8,3,7]
print(containerMaxWater(arr))  # ✅ Output: 49
```

# BFS For Minimum Value in the Tree

```python
from collections import deque

class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value=value
        self.left=left
        self.right=right
        
def findMinBFS(root):
        if not root:
            return None
        queue = deque([root])
        min_val = root.value

        while queue:
            
            node = queue.popleft()

            if node.value < min_val:
                min_val = node.value
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        return min_val

# Construct the tree
#        10
#       /  \
#     5     15
#    / \    / \
#   2  7   12  20

root = TreeNode(10)
root.left = TreeNode(5, TreeNode(2), TreeNode(7))
root.right = TreeNode(15, TreeNode(12), TreeNode(20))

print(findMinBFS(root))  # Output → 2
```

# BFS MaxDepth

```python
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value=value
        self.left=left
        self.right=right

def bfsMaxDepth(root):
    if not root:
        return []

    queue = deque([root])
    depth = 0
    
    while queue:
        for i in range(len(queue)):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        depth += 1
    return depth
                
# Construct the tree
#        10
#       /  \
#     5     15
#    / \    / \
#   2  7   12  20

root = TreeNode(10)
root.left = TreeNode(5, TreeNode(2), TreeNode(7))
root.right = TreeNode(15, TreeNode(12), TreeNode(20))



print(bfsMaxDepth(root))  # Output → 2
```

# Invert A TREE

```python
class TreeNode:
    def __init__(self, value=0, left=None, right=None):
        self.value=value
        self.left=left
        self.right=right

def invertTree(root):
    if not root:
        return None
    right = root.right
    left = root.left
    root.right = left
    root.left = right
    
    invertTree(root.right)
    invertTree(root.left)
    return root
    
    

# Construct the tree
#        10
#       /  \
#     5     15
#    / \    / \
#   2  7   12  20

root = TreeNode(10)
root.left = TreeNode(5, TreeNode(2), TreeNode(7))
root.right = TreeNode(15, TreeNode(12), TreeNode(20))



print(invertTree(root))  # Output → 2
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def isSameTree(s : TreeNode, t : TreeNode):
    if not s and not t:
        return True

    if s and t and s.val == t.val:
        return (isSameTree(s.left, t.left) and isSameTree(s.right, t.right))
    return False


# Build two trees
root1 = TreeNode(10, TreeNode(5), TreeNode(15))
root2 = TreeNode(10, TreeNode(5), TreeNode(15))
root3 = TreeNode(10, TreeNode(6), TreeNode(15))

print(isSameTree(root1, root2))  # ✅ True
print(isSameTree(root1, root3))  # ❌ False
```

# Binary Tree Level Order Traversal 

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def bfsLOT(root):
    if not root:
        return []
    
    queue = deque([root])
    result = []

    # Level-order traversal
    while queue:
        level = []  # store nodes of the current level
        
        for _ in range(len(queue)):  # process all nodes in the current level
            node = queue.popleft()
            level.append(node.val)  # store the node value
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        result.append(level)  # after finishing the level, add to result
    
    return result


# Construct the tree
#        10
#       /  \
#     5     15
#    / \    / \
#   2  7   12  20

root = TreeNode(10)
root.left = TreeNode(5, TreeNode(2), TreeNode(7))
root.right = TreeNode(15, TreeNode(12), TreeNode(20))

print(bfsLOT(root))
```

# Inorder Traversal(DFS)

**it will first return the the root then it will go to the left then right dfsInorder(root.left) + [root.val] + dfsInorder(root.right)**

**left subtree ➡️ root ➡️ right subtree**

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val=val
        self.left=left
        self.right=right


def dfsInorder(root):
    if not root:
        return []
        
    return dfsInorder(root.left) + [root.val] + dfsInorder(root.right)

# Example usage
root = TreeNode(10)
root.left = TreeNode(5, TreeNode(2), TreeNode(7))
root.right = TreeNode(15, TreeNode(12), TreeNode(20))

print(dfsInorder(root))
```

# Preorder Traversal(DFS)

**it will first return the the root then it will go to the left then right[root.val] + dfsInorder(root.left) + dfsInorder(root.right)**

**root ➡️ left subtree ➡️ right subtree**

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val=val
        self.left=left
        self.right=right


def dfsInorder(root):
    if not root:
        return []
        
    return [root.val] + dfsInorder(root.left) + dfsInorder(root.right)

# Example usage
root = TreeNode(3)
root.left = TreeNode(9)
root.right = TreeNode(20, TreeNode(15), TreeNode(7))

print(dfsInorder(root))
```

# Valid Binary Search Tree(recursve)


```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val=val
        self.left=left
        self.right=right

def valid_BST(root, left=float('-inf'), right=float('inf')):
    if not root:
        return True

    if not (left < root.val < right):
        return False

    return (valid_BST(root.left, left, root.val) and 
            valid_BST(root.right, root.val, right))

# Example usage
root = TreeNode(10)
root.left = TreeNode(11)
root.right = TreeNode(14, TreeNode(12), TreeNode(15))

print(valid_BST(root))  # Output: False
```

# Kth smallest element in a Tree

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val=val
        self.left=left
        self.right=right

def inorder_traversal(root):
    if not root:
        return []
    return inorder_traversal(root.left) + [root.val] + inorder_traversal(root.right)
   
def kth_element(root, k):
    arr = inorder_traversal(root)
    # k-th smallest is at index k-1
    return arr[k-1] 

# Example usage
root = TreeNode(10)
root.left = TreeNode(5, TreeNode(2), TreeNode(7))
root.right = TreeNode(15, TreeNode(12), TreeNode(20))
print(inorder_traversal(root))
print(kth_element(root, 2))
```

# House Robber I

```python
def robHouse(arr):
    rob1 = 0
    rob2 = 0

    for n in arr:
        temp = max(rob1 + n, rob2)
        rob1 = rob2
        rob2 = temp
    return rob2

arr = [1,2,3,4,5]
robHouse(arr)
```

# House Robber II

```python
def robHouseII(arr):
    rob1 = 0
    rob2 = 0
    rob4 = 0
    rob3 = 0

    for i in range(len(arr) - 1):
        temp = max(rob1 + arr[i], rob2)
        rob1 = rob2
        rob2 = temp

    for i in range(1, len(arr)):
        temp = max(rob3 + arr[i], rob4)
        rob3 = rob4
        rob4 = temp

    
    return max(arr[0],rob2, rob4)

arr = [1]
robHouseII(arr)
```

# Heapify from scratch (MaxHeap)

```python
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    # If left child is larger than root
    if left < n and arr[left] > arr[largest]:
        largest = left

    # If right child is larger than largest so far
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not root, swap and continue heapifying
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def build_max_heap(arr):
    n = len(arr)
    # Build heap (rearrange array)
    for i in range(n//2 - 1, -1, -1):  # from last non-leaf to root
        heapify(arr, n, i)

arr = [3, 9, 2, 1, 4, 5]
build_max_heap(arr)
print("Max Heap:", arr)
```

```python
def heap_sort(arr):
    n = len(arr)
    build_max_heap(arr)
    
    for i in range(n - 1, 0, -1):
        # Move current root to end
        arr[i], arr[0] = arr[0], arr[i]
        # Call heapify on the reduced heap
        heapify(arr, i, 0)
    return arr

arr = [3, 9, 2, 1, 4, 5]
print("Sorted:", heap_sort(arr))
```

# Top K numbers

```python
import heapq
from collections import Counter

def frequencyCounter(nums, k):
    freq = Counter(nums)
    heap = []

    for num, count in freq.items():
        heapq.heappush(heap, (count, num))
       
        
    res = [heapq.heappop(heap)[1] for _ in range(k)]
    return res

# Example
nums = [1,1,1,2,2,2,3]
k = 2
print(frequencyCounter(nums, k))  # Output: [1,2]
```

# Built In heap Sort

```python
def heapSort(arr):
    heapq.heapify(arr)
    n = len(arr)
    new_list = [0] * n

    for i in range(n):
        minn = heapq.heappop(arr)
        new_list[i] = minn
    return new_list

heapSort([1,2,3,4,5,6,8,6,5,110,23])
```

# Reverse A linked List

```python
# Define ListNode
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Reverse linked list
def reverse_a_linked_list(head: ListNode):
    prev, curr = None, head
    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt
    return prev

# Helper to print linked list
def print_linked_list(head: ListNode):
    curr = head
    while curr:
        print(curr.val, end=" -> " if curr.next else "")
        curr = curr.next
    print()

# Create linked list: 1 -> 2 -> 3 -> 4 -> 5
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))

print("Original linked list:")
print_linked_list(head)

# Reverse
reversed_head = reverse_a_linked_list(head)

print("Reversed linked list:")
print_linked_list(reversed_head)
```

# Linked List cycle

```python
class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None


def linked_cycle(head):
    slow, fast = head, head

    while fast and fast.next:
        slow = slow.next          # move slow by 1
        fast = fast.next.next     # move fast by 2

        if fast == slow:          # if they meet, there’s a cycle
            return True

    return False  # if fast reaches None → no cycle


# -------- Test the function --------
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Cycle present?", linked_cycle(head))  # ❌ No cycle

# Create a cycle: connect node 5 → node 3
head.next.next.next.next.next = head.next.next

print("Cycle present?", linked_cycle(head))  # ✅ Cycle detected
```

# Merge Two Sorted Linked List

```python
class ListNode:
    def __init__(self,value):
        self.value=value
        self.next=None

def mergeTwoLists(list1, list2):
    dummy = ListNode(0)
    tail = dummy

    while list1 and list2:
        if list1.value < list2.value:
            tail.next = list1
            list1 = list1.next
        else:
            tail.next = list2
            list2 = list2.next
        
        tail = tail.next

    if list1:
        tail.next = list1
    elif list2:
        tail.next = list2
    return dummy.next

# -------- Test the function --------
# List 1: 1 -> 3 -> 5
l1 = ListNode(1)
l1.next = ListNode(3)
l1.next.next = ListNode(5)

# List 2: 2 -> 4 -> 6
l2 = ListNode(2)
l2.next = ListNode(4)
l2.next.next = ListNode(6)


# Merge
merged = mergeTwoLists(l1, l2)

# Print merged list
while merged:
    print(merged.value, end=" -> ")
    if merged.next == None:
        print("Null")
    merged = merged.next
```
