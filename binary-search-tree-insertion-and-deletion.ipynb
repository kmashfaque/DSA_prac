```python
class Node:
    def __init__(self, data):
        self.data = data
        self.right = None
        self.left = None

def insert(node, data):
    if node is None:
        return Node(data)

    elif data < node.data:
        node.left = insert(node.left, data)
        
    elif data > node.data:
        node.right = insert(node.right, data)
        
    else:
        print("data already exist")

    return node

def printNode(node):
    if node is not None:
        printNode(node.left)
        print(node.data)
        printNode(node.right)

def minValueNode(node):
    current = node

    while current.left is not None:
        current = current.left
    return current

def deleteNode(root, data):
    if root is None:
        return root

    # Step 1: Traverse to find the node
    if data < root.data:
        root.left = deleteNode(root.left, data)
    elif data > root.data:
        root.right = deleteNode(root.right, data)
    else:
         # Step 2: Node found â€” handle the three cases
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp

        # Case 2: Node with two children
        # Find inorder successor (smallest in right subtree)
        temp = minValueNode(root.right)
        root.data = temp.data   # Replace data
        # Delete the inorder successor
        root.right = deleteNode(root.right, temp.data)

    return root
        
# node = Node(5)
# insert(node, 3)
# insert(node, 7)
# insert(node, 1)
# insert(node, 2)
# insert(node, 4)
# insert(node, 6)
# insert(node, 8)
# printNode(node)

# Example usage:
root = Node(5)
for val in [3, 7, 2, 4, 6, 8]:
    insert(root, val)

print("Original Tree:")
printNode(root)

# root = deleteNode(root, 2)  # Example delete
# print("\nAfter deleting 2:")
# printNode(root)

# Output : 2,3,4,5,6,7,8
```

```python
# find a value from the BST

def findVal(node, val):
    
    if node is None:
        print("Value is Not found")
        return None

    elif node.data == val:
        print("The value found ==>", node.data)

    elif val < node.data:
        return findVal(node.left, val)

    elif val > node.data:
        return findVal(node.right, val)

    
    
findVal(root, 5)
```

```python
# finding max depth of a BST tree
from collections import deque

def maxdepth(root):

    if root is None:
        return root
        
    queue = deque([root])
    max_depth = 0

    while queue:
        
        for _ in range(len(queue)):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        max_depth += 1
    return max_depth

maxdepth(root)
```

```python
# find minimum in a Tree using Breadth First Search

def BFSmin(root):

    if root is None:
        return None

    queue = deque([root])
    min_val = root.data

    while queue:
        node = queue.popleft()
        if min_val > node.data:
            min_val = node.data
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return min_val

BFSmin(root)
```
